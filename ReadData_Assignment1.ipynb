```python
import wget
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from zipfile import ZipFile
```

```python
for i in range(0,3):
    url = f'https://data.binance.vision/data/spot/daily/trades/BTCUSDT/BTCUSDT-trades-2022-03-0{i+1}.zip'
    wget.download(url)
    
    with ZipFile(f'BTCUSDT-trades-2022-03-0{i+1}.zip', 'r') as zip:
    
        # printing all the contents of the zip file
        zip.printdir()    

        # extracting all the files
        print('Extracting all the files now...')
        zip.extractall()
        print('Done!')
```

```python
data1 = pd.read_csv('BTCUSDT-trades-2022-03-01.csv')
data2 = pd.read_csv('BTCUSDT-trades-2022-03-02.csv')
data3 = pd.read_csv('BTCUSDT-trades-2022-03-03.csv')
```

```python
df1 = data1.copy()
df2 = data2.copy()
df3 = data3.copy()
```

###### In the downloaded dataset, actual tick data is set as column names, so here we are adding the column names of df2, as a last data row in df1, and adding the column names of df3, as a last data row in df2

```python
df1.loc[len(df1.index)] = df2.columns
df2.loc[len(df2.index)] = df3.columns
```

###### Now, we need to club the all 3 datasets into one, so we need to reset their column names as same for all 3 dataFrames

```python
df1.rename(columns={'1274557543':'Trade_Id', '43160.00000000':'Price', '0.00056000':'Qty', '24.16960000':'QuoteQty',
                    '1646092800000':'time','True':'isBuyerMaker', 'True.1':'isBestMatch'}, inplace=True)
df2.rename(columns={'1276424414':'Trade_Id', '44421.20000000':'Price', '0.00097000':'Qty', '43.08856400':'QuoteQty',
                    '1646179200000':'time','True':'isBuyerMaker', 'True.1':'isBestMatch'}, inplace=True)
df3.rename(columns={'1278077545':'Trade_Id', '43892.99000000':'Price', '0.00158000':'Qty', '69.35092420':'QuoteQty',
                    '1646265600000':'time','False':'isBuyerMaker', 'True':'isBestMatch'}, inplace=True)
```

###### Now here we combine the data of 3 days into 1 dataset

```python
df4 = pd.concat([df1, df2, df3])
df4
```

```python
print(df4.shape)
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

###### Resetting the index

```python
df4['Index'] = range(0, len(df4))
df4.set_index('Index', inplace=True)
df4
```

###### Making a copy of our final dataset for further analysis

```python
data = df4.copy()  #Final Dataset for further Analysis
data.to_csv('BTCUSDT_DataSet.csv')
```
